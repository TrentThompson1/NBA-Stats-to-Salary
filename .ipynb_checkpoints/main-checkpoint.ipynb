## Project Overview

Our project uses data from [Basketball-Reference](https://www.basketball-reference.com/) to study the relationship between NBA player performance and salary. We started with a simple idea: compare a player's pay to their box-score stats (points, rebounds, assists, steals, blocks). From there, we refined our goal into a more analytical question:

This lets us look at:
- Which stats matter most for getting paid
- Which positions and roles earn more
- And which contracts provide the most "value" compared to a player's production


### Q1: What stats are most strongly related to NBA salary?

We first computed a correlation matrix between salary and basic box-score stats (points, assists, steals, blocks, turnovers, games, FG%, 3P%, and rebounds), and visualized it with a heatmap.

### A1: Analysis

- We merged per-game stats with current salary from Basketball-Reference.
- We computed the correlation between 'Salary' and key Statistics.
- We plotted a heatmao and several scatterplots:
      - Salary vs Points
      - Salary vs Assists
      - Salary vs Age
      - Average salary and average points by position

### C1: Conclusion

- **Points per game** has the strongest correlation with salary, followed by **assists**. This matches basketball intuition: scoring and creating offense drive pay.
- **By position**, point guards have both the highest **average salary** and the highest **average points per game**, which is consistant with the idea that primary ball handlers and scorers get paid the most.
- **Field goal percentage** and **3-point percentage** show much weaker correlation with salary. This result may seem suprising at first, because modern NBA offenses place a high on efficient shooting, especially 3-point shots. However, our dataset includes **every player who logged minutes**, not just rotation players. this means players who appear in only a few games can end up with **extremely high or low shooting percentages** due to the very small sample size.

**Merging, and cleaning the Baskeball Data**

```python
import pandas as pd

stats = pd.read_html("https://www.basketball-reference.com/leagues/NBA_2025_per_game.html")[0]
#pulling both stats and salary from bball ref
salary = pd.read_html("https://www.basketball-reference.com/contracts/players.html", header=[0,1])[0]
#flatten column names dropping unnamed filler labels
salary.columns = ['_'.join([str(c) for c in col if c and c != 'Unnamed']) for col in salary.columns]
#remove repeated headers rows that appear in table body. rename tm to team.
stats = stats[~stats['Player'].str.contains('Player', na=False)]
stats = stats.rename(columns={'Tm':'Team'})
#flatten contracts to simple names
salary = salary.rename(columns={'Unnamed: 1_level_0_Player':'Player', 'Unnamed: 2_level_0_Tm':'Team', 'Salary_2025-26':'Salary'})
salary = salary[['Player','Team','Salary']].copy()
# clean and merge
salary['Salary'] = (
    salary['Salary']
      .replace('[\$,]', '', regex=True)   # remove $ and commas                      # convert to numeric
)
#inner join by keeping player that only appear in both game stats and salary
merged = pd.merge(stats, salary[['Player','Salary']], on='Player', how='inner')
merged['Salary'] = merged['Salary'].astype('int')

merged = merged.sort_values(by=['Team'])
#dropping dups if they exist
merged = merged.drop_duplicates(subset=['Player', 'Age'], keep='first').reset_index(drop=True)
merged = merged.sort_values(by=['Rk'])
```

### Q2: How well can we predict a player's salary using only basic box-score statistics and age?

We trained a **Linear regression model** to predict each player's current salary using only the following features:
- Points per game (PTS)
- Assists per game (AST)
- Rebounds per game (TRB)
- Steals per game (STL)
- Blocks per game (BLK)
- Turnovers per game (TOV)
- Age

These are simply, widely available statistics, so the question is: *How much of NBA salary can we explain using just this limited information?*

### A2: Analysis (Linear Regression)

We split the dataset into training (80%) and testing (20%) sets and fit a standard linear regression model.

The model performance on the unseen test data was:

- **R = 0.70**
- **MAE  \$6.7 million**

**Interpretation:**
- An R of 0.70 means that about 70% of the variation in NBA salaries can be explained using these 7 basic variables
- The MAE of about 6.7 million indicates the typical prediction error.


### C2: Conclusion
- Even with very limited box-score data, salary prediction is pretty strong, confirming that scoring and offensive impact (PTS, AST) drive most NBA contract value.
- The remaining 30% of unexplained variation likely comes from:
    - Previous seasons rather than just the current year
    - Awards (All-NBA, MVP, DPOY)
    - Team market size and cap space
    - Contract timing (rookie deals, supermax extensions, veteran minimums)

These results lead naturally into **Cycle 2**, where we investigate deeper questions such as:
- Why shooting percentages (FG%, 3PT%) do not strongly correlate with salary
- whether low-minute players distort certain statistics
- And whether different types of players (clusters) are paid differently

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
#predictive model guessing salary to their production
#feature set for salary prediction(box score and stats)
x_features = ['PTS', 'AST', 'TRB', 'STL', 'BLK', 'Age', 'TOV']

X = merged[x_features].fillna(0)
y = merged['Salary']
#test train split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2)
#linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)
#predict
y_hat = model.predict(X_test)


print("R^2:", round(r2_score(Y_test, y_hat), 2))
print("MAE:", round(mean_absolute_error(Y_test, y_hat), 3))

#showing the the predicted and actual salary in a scatterplot with an ideal line
plt.figure(figsize=(12,10))
plt.scatter(Y_test, y_hat, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Predicted vs Actual NBA Salaries")
plt.tight_layout()
plt.show()
```

### Q3: Do low-minute players distort the relationship between shooting percentages and salary?

In Cycle 1, we saw that FG% and 3P%, had suprisingly weak correlations with salary.
One hypothesis was that our dataset includes may low-minute players whose percentages are inflated by small sample sizes.

To test this, we filtered the data to include only players with:
- at least 20 games played
- at least 15 minutes per game

We then recomputed the salary correlations.

### A3: Analysis

Even after filtering out 133 low-minute players, the correlation changed very little:

| FG%    | 0.08  0.09 |
| 3P%    | 0.12  0.13 |

### C3: Conclusion

Filtering out small-sample players did not meaningfully increase the relationship between shooting percentages and salary.
This suggests that:

**NBA teams do not base player salaries on shooting efficiency along.**
Instead, they reward:
- Offensive volume (PTS & AST)
- Usage and role
- Track record and awards
- Contract timing and market factors

This motivates us to look beyond individual stats and instead example **player archetypes** using clustering.

```python
min_games = 20
min_minutes = 15

df_filt = merged[(merged['G'] >= min_games) & (merged['MP'] >= min_minutes)]
df_filt.head()
```

### Q4: What player archetypes exist in the NBA based on performance and salary

Do distinct types of players exist: Clusters of players who are paid similarly because their roles and statistical profiles are similar?

To investigate this, we performed K-Means clustering using six core performance variables:
- PTS
- AST
- TRB
- STL
- BLK
- TOV

we then used the Elbow Method, which indicated that k = 5 clusters was an appropriate choice. Finally, we computed the average stats and salaries per cluster and visualized the groups using radar plots.

### A4: Analysis

The clusters revealed five clear groups of players, each representing a different role.
#### Cluster 0 - Deep Bench
- 4.6 PTS, 1 AST, 2 REB
- Salary: $4.1M

Players who rarely see significant minutes(rookies, two-way contracts, fringe rotation pieces)

#### Cluster 1 - Primary Stars
- 22 PTS, 6 AST, 6 REB
- Highest turnovers (ball-dominant)
- Salary: $37.2M

These players carry the offense and initiate plays. Radar plots show largest values across most categories

#### Cluster 2 - Secondary Scorers
- 13 PTS, 3-4 AST, 3-4 REB
- solid steals
- Salary: $16.3M

Strong contributors who support the offense but are not the primary option.

#### Cluster 3 - Defensive Bigs
- 13 PTS, 8.5 REB, 1.4 BLK
- Low assists and steals
- Salary: $19.2M

These are interior players whose impace comes from rebounding and rim protection.

#### Cluster 4 - Rotation players
- 8.6 PTS, 4.3 REB, 1.7 AST
- Salary $9.2M

Bench or low-usage players who contribute in limited roles

### C4: Conclusion
The cluster analysis provides several important insights:
1. NBA salaries map extremely well onto player archetypes.
2. Player roles, not individual stats, drive contract value
3. This confirms that NBA teams pay for roles, opportunity, and offensive responsibility, not just raw stats

This second data-science cycle deepened our findings by moving from single-variable relationships to understanding how whole groups of players differ in performance and pay.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

cluster_features = ['PTS', 'AST', 'TRB', 'STL', 'BLK', 'TOV']

X_cluster = merged[cluster_features].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)
```

```python
inertias = []
ks = range(2, 20)

for k in ks:
    km = KMeans(n_clusters=k, random_state = 42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(8,5))
plt.plot(ks, inertias, 'o-')
plt.title("Elbow method for Optimal K")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()
```

```python
k = 5 # steady smaller drops observed after k = 5
kmeans = KMeans(n_clusters = k, random_state=42, n_init = 10)
merged['Cluster'] = kmeans.fit_predict(X_scaled)
num_cols = cluster_features + ['Salary']

# Force them to numeric, coerce anything weird to NaN
for col in num_cols:
    merged[col] = pd.to_numeric(merged[col], errors='coerce')

cluster_summary = merged.groupby('Cluster')[cluster_features + ['Salary']].mean().round(2)
cluster_summary
```

```python
cluster_stats = cluster_summary[['PTS','AST','TRB','STL','BLK','TOV']]
labels = cluster_stats.columns.tolist()

cluster_norm = (cluster_stats - cluster_stats.min()) / (cluster_stats.max() - cluster_stats.min())

def make_radar(row, cluster_id):
    values = row.values.flatten().tolist()
    values += values[:1]

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

    ax.plot(angles, values, linewidth=2)
    ax.fill(angles, values, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=12)
    ax.set_title(f"Radar Plot for Cluster {cluster_id}", fontsize=16, pad=20)

    plt.show()


for cid in cluster_norm.index:
    make_radar(cluster_norm.loc[cid], cid)
```

### Q5: Does running separate regression models inside each cluster improve salary prediction?
In cycle 1, we used a single linear regression model across all NBA players and achieved:
- R  0.70
- MAE  $6.7M

However, the NBA market is not uniform. Superstars, defensive specialists, bench players, and high-minute starters all get paid according to different norms.

To test whether segmentation imporves prediction accuracy, we ran separate regression models whithin each K-Means cluster created earlier. These clusters represent distinct player archetypes based on production.

### A5: Analysis

The regression models showed dramatically different performance depending on the cluster:

#### Cluster 4: The only group where stats actually predict salary
Cluster 4, which included high-minute rotation starters, had:
- R = 0.368
- MAE = $4.47M

These players have consistent production, predictable roles, and market-level contracts, so linear regression captures a real salary relationship.

### C5: Conclusion
Running regression inside each cluster does not improve predictive accuracy compared to the global model.
In fact, most cluster-specific models perform worse, which reveals an important insight about the NBA salary system:

NBA salaries are not determined by the same statistical factors for every type of player.
- Stars - paid for reputation, branding, and contract timing
- Bench players - paid at a fixed market floor
- Defensive bigs - valued for skills not captured fully in standard box scores
- High-minute starters - salary is strongly linked to production

Clustering does not improve regression accuracy, but it explains why the global model works imperfectly, different player archetypes follow different salary rules.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

cluster_results = {}

for c in sorted(merged['Cluster'].unique()):
    # take only this cluster
    df_c = merged[merged['Cluster'] == c].copy()
    
    # keep only rows with complete data in features + Salary
    subset = df_c[cluster_features + ['Salary']].dropna()
    
    X_c = subset[cluster_features]
    y_c = subset['Salary']
    
    n = len(subset)
    # skip tiny clusters
    if n < 8:
        cluster_results[c] = {"R2": None, "MAE": None, "n_players": n}
        continue
    
    X_train, X_test, y_train, y_test = train_test_split(
        X_c, y_c, test_size=0.2, random_state=42
    )
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    cluster_results[c] = {
        "R2": round(r2_score(y_test, y_pred), 3),
        "MAE": round(mean_absolute_error(y_test, y_pred), 0),
        "n_players": n
    }

cluster_results
```

## Final Conclusion

Our analysis of NBA salaries reveals that player compensation is driven far more by offensive responsibility and role than by efficiency or isolated box score metrics. Across the league, points and assists consistently showed the strongest relationships to salary, confirming that teams pay a premium for players who create offense either through scoring or playmaking. In contrast, shooting efficiency (FG% and 3P%) and defensive statistics had surprisingly weak correlations with pay, and even removing low-minute players did little to change this resultdemonstrating that efficiency alone does not meaningfully influence contract value for most players. Position-based analysis also supported this conclusion: the highest-paid positions were those with the largest offensive load, not necessarily the most efficient or defensively impactful ones. Clustering deepened these insights by showing that different types of players follow different economic patterns. Bench and low-usage players earn near-fixed minimum contracts regardless of production, while superstars command near-max salaries driven by market rules and reputation rather than linear relationships with stats. Defensive-oriented centers likewise showed salary patterns not strongly tied to traditional box score measures. Only high-minute, high-production starters exhibited a predictable relationship between performance and pay, explaining why the global regression model performed moderately well even though cluster-specific models often did not. Taken together, our findings demonstrate that NBA salaries reflect a players role, usage, and perceived offensive value more than their efficiency or isolated contributions, highlighting the complexity and segmentation of the leagues labor market.
